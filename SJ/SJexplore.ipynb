```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
kaggleHouse = pd.read_csv("train.csv")
```

```python
df = kaggleHouse
featureTypes = df.columns
```

```python
featureTypes
```

```python
#Automated dictionary creation for type of variable:

typeConvertDict = {}

for feature in featureTypes:

    print(feature)

    sns.catplot(data = df, x = feature, kind='count')

    plt.show()

    #df[feature[0]].value_counts(dropna = False).plot( kind = 'bar')

    currentSuggestion = str(input(prompt = "Ordinal, Categorical, Float, Integer, or Boolean?"))

    typeConvertDict[feature] = currentSuggestion



import pickle



f = open("typeConvertDict.pkl","wb")

pickle.dump(typeConvertDict,f)

f.close()
```
